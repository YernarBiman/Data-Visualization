```python
from sklearn import datasets
import json
import umap
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x, y = iris["data"], iris["target"]
```

```python
reducer = umap.UMAP()
```

```python
embedding = reducer.fit_transform(x)
embedding.shape
```

```python
data = pd.DataFrame(embedding, columns=["Comp1","Comp2"])
data["Label"] = y
```

```python
data
```

```python
# TODO add the Target Information into  your json file
with open('data_umap.json', 'w') as f:
    json.dump(np.array(data).tolist(), f)
```

```python
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

sns.lmplot(x='Comp1',
           y='Comp2',
           data=data,
           fit_reg=False,
           legend=True,
           size=9,
           hue='Label',
           scatter_kws={"s":200, "alpha":0.3}
          )

plt.title("UMAP", weight='bold').set_fontsize('14')
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
```
